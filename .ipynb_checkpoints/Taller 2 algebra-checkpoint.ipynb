```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar tus datos en un DataFrame de pandas (reemplaza 'data.csv' con tu archivo de datos)
data = pd.read_csv('dataset.csv')
columnas_deseadas = ['assets', 'profits','roa','sales']
nuevo_data = data[columnas_deseadas]
nuevo_data.info()
```

```python
# Calcular medias de profits y sales
mean_profits = nuevo_data["profits"].mean()
mean_sales = nuevo_data["sales"].mean()
```

```python
# Crear el gráfico de dispersión
plt.figure(figsize=(8, 6))
plt.scatter(nuevo_data["profits"], nuevo_data["sales"], label="Datos")
plt.axhline(y=mean_sales, color='r', linestyle='-', label=f"Media de Sales: {mean_sales:.2f}")
plt.axvline(x=mean_profits, color='g', linestyle='-', label=f"Media de Profits: {mean_profits:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

 

# Crear una matriz de gráficos de dispersión
sns.set(style="ticks")
sns.pairplot(nuevo_data, vars=["profits", "assets", "roa", "sales"])

 

# Títulos personalizados para cada gráfico
plt.subplots_adjust(top=0.95)
plt.suptitle("Matriz de Gráficos de Dispersión: Profits vs Assets, ROA, Sales", fontsize=16)

 

# Mostrar la matriz de gráficos de dispersión
plt.show()
```
